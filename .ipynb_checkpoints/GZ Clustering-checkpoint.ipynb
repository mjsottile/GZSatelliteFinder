```python
from sklearn import metrics, cluster
from sklearn.preprocessing import StandardScaler
import os
import csv
import numpy as np
```

```python
def readone(fname):
    f = open(fname,'r')
    dat = csv.reader(f)
    for l in dat:
        return l[0],map(float,l[1:])
```

```python
path = 'measurements/'
listing = os.listdir(path)
data = []
labels = []
for infile in listing:
    d = readone("measurements/"+infile)
    if not (d == None):
        lbl,dat = d
        labels.append(lbl)
        data.append(dat)
x = np.array(data)
```

```python
af=cluster.AffinityPropagation().fit(x)
af.labels_
```

```python
s_x = StandardScaler().fit_transform(x)
db=cluster.DBSCAN(eps=0.2).fit(s_x)
db.labels_
```

```python
f = open("foo.html",'w')
for i in range(0,max(af.labels_)+1):
    f.write("<H1>"+str(i)+"</H1>\n")
    for j in range(0,len(labels)):
        if af.labels_[j] == i:
            f.write("<IMG TEXT=\""+labels[j]+"\" SRC=\"images/"+labels[j]+".jpg\">\n")
    f.write("<P>\n")
f.close()
```
